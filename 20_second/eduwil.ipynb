```python
df = pd.read_excel('score.xlsx', index_col=[0], header=[0,1])

eduwil1 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "에듀윌") & (df["Unnamed: 2_level_0", "회차"] ==1)]
eduwil1.style.background_gradient(cmap='summer_r')
# print(eduwil1)
# print(eduwil1)

eduwil2 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "에듀윌") & (df["Unnamed: 2_level_0", "회차"] ==2)]
eduwil2.style.background_gradient(cmap='summer_r')

eduwil3 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "에듀윌") & (df["Unnamed: 2_level_0", "회차"] ==3)]
eduwil3.style.background_gradient(cmap='summer_r')

eduwil4 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "에듀윌") & (df["Unnamed: 2_level_0", "회차"] ==4)]
eduwil4.style.background_gradient(cmap='summer_r')

display(eduwil1.style.background_gradient(cmap='summer_r'))
# display(eduwil2.style.background_gradient(cmap='summer_r'))
# display(eduwil3.style.background_gradient(cmap='summer_r'))
# display(eduwil4.style.background_gradient(cmap='summer_r'))
# 각 과목당 오답률
# df.iloc[:,9:]
```

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
eduwil1_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="에듀윌1_못푼개수", cmap="rainbow")
eduwil1_total.plot(kind="bar", legend=True, ax=ax2, title="에듀윌1_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()


# figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# figure.set_size_inches(15,4)
# eduwil2_nonreplied.plot(kind="bar", legend=False, ax=ax1, title="에듀윌2_못푼개수", cmap="rainbow")
# eduwil2_total.plot(kind="bar", legend=True, ax=ax2, title="에듀윌2_못푼개수+틀린개수", cmap="rainbow")
# for p in ax1.patches:
#     ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# for q in ax2.patches :
#     ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
# plt.show()


# figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# figure.set_size_inches(15,4)
# eduwil3_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="에듀윌3_못푼개수", cmap="rainbow")
# eduwil3_total.plot(kind="bar", legend=True, ax=ax2, title="에듀윌3_못푼개수+틀린개수", cmap="rainbow")
# for p in ax1.patches:
#     ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# for q in ax2.patches :
#     ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
# plt.show()

# figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# figure.set_size_inches(15,4)
# eduwil4_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="에듀윌4_못푼개수", cmap="rainbow")
# eduwil4_total.plot(kind="bar", legend=True, ax=ax2, title="에듀윌4_못푼개수+틀린개수", cmap="rainbow")
# for p in ax1.patches:
#     ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# for q in ax2.patches :
#     ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
# plt.show()
```
