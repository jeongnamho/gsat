```python
df = pd.read_excel('score1.xlsx', index_col=[0], header=[0,1])

letuin1 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "렛유인_파이널") & (df["Unnamed: 2_level_0", "회차"] ==1)]
letuin1.style.background_gradient(cmap='summer_r')
# print(letuin1)
# print(letuin1)

letuin2 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "렛유인_파이널") & (df["Unnamed: 2_level_0", "회차"] ==2)]
letuin2.style.background_gradient(cmap='summer_r')

letuin3 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "렛유인_파이널") & (df["Unnamed: 2_level_0", "회차"] ==3)]
letuin3.style.background_gradient(cmap='summer_r')

display(letuin1.style.background_gradient(cmap='summer_r'))
display(letuin2.style.background_gradient(cmap='summer_r'))
display(letuin3.style.background_gradient(cmap='summer_r'))
# 각 과목당 오답률
# df.iloc[:,9:]
```

```python
letuin1_nonreplied = letuin1.iloc[:,[2,4,6,8]]
letuin1_total = letuin1.iloc[:,[3,5,7,9]]
letuin2_nonreplied = letuin2.iloc[:,[2,4,6,8]]
letuin2_total = letuin2.iloc[:,[3,5,7,9]]
letuin3_nonreplied = letuin3.iloc[:,[2,4,6,8]]
letuin3_total = letuin3.iloc[:,[3,5,7,9]]
```

### 개인별  (못푼 개수) & (못푼+틀린 개수) 

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
letuin1_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="렛유인1_못푼개수", cmap="rainbow")
letuin1_total.plot(kind="bar", legend=True, ax=ax2, title="렛유인1_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()


# figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# figure.set_size_inches(15,4)
# letuin2_nonreplied.plot(kind="bar", legend=False, ax=ax1, title="렛유인2_못푼개수", cmap="rainbow")
# letuin2_total.plot(kind="bar", legend=True, ax=ax2, title="렛유인2_못푼개수+틀린개수", cmap="rainbow")
# for p in ax1.patches:
#     ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# for q in ax2.patches :
#     ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
# plt.show()


# figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# figure.set_size_inches(15,4)
# letuin3_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="렛유인3_못푼개수", cmap="rainbow")
# letuin3_total.plot(kind="bar", legend=True, ax=ax2, title="렛유인3_못푼개수+틀린개수", cmap="rainbow")
# for p in ax1.patches:
#     ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# for q in ax2.patches :
#     ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
# plt.show()
```
