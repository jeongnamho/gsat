# <font color=red> hackers_real </font>

```python
df = pd.read_excel('score.xlsx', index_col=[0], header=[0,1])

hackers_real1 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "해커스실전") & (df["Unnamed: 2_level_0", "회차"] ==1)]
hackers_real1.style.background_gradient(cmap='summer_r')
# print(hackers_real1)
# print(hackers_real1)

hackers_real2 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "해커스실전") & (df["Unnamed: 2_level_0", "회차"] ==2)]
hackers_real2.style.background_gradient(cmap='summer_r')

hackers_real3 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "해커스실전") & (df["Unnamed: 2_level_0", "회차"] ==3)]
hackers_real3.style.background_gradient(cmap='summer_r')

hackers_real4 = df.loc[(df["Unnamed: 1_level_0", "교재"] == "해커스실전") & (df["Unnamed: 2_level_0", "회차"] ==4)]
hackers_real4.style.background_gradient(cmap='summer_r')

display(hackers_real1.style.background_gradient(cmap='summer_r'))
display(hackers_real2.style.background_gradient(cmap='summer_r'))
display(hackers_real3.style.background_gradient(cmap='summer_r'))
display(hackers_real4.style.background_gradient(cmap='summer_r'))
# 각 과목당 오답률
# df.iloc[:,9:]
```

```python
hackers_real1_nonreplied = hackers_real1.iloc[:,[2,4,6,8]]
hackers_real1_total = hackers_real1.iloc[:,[3,5,7,9]]
hackers_real2_nonreplied = hackers_real2.iloc[:,[2,4,6,8]]
hackers_real2_total = hackers_real2.iloc[:,[3,5,7,9]]
hackers_real3_nonreplied = hackers_real3.iloc[:,[2,4,6,8]]
hackers_real3_total = hackers_real3.iloc[:,[3,5,7,9]]
hackers_real4_nonreplied = hackers_real4.iloc[:,[2,4,6,8]]
hackers_real4_total = hackers_real4.iloc[:,[3,5,7,9]]
```

### 개인별  (못푼 개수) & (못푼+틀린 개수) 

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
hackers_real1_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="해커스실전1_못푼개수", cmap="rainbow")
hackers_real1_total.plot(kind="bar", legend=True, ax=ax2, title="해커스실전1_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()


figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
hackers_real2_nonreplied.plot(kind="bar", legend=False, ax=ax1, title="해커스실전2_못푼개수", cmap="rainbow")
hackers_real2_total.plot(kind="bar", legend=True, ax=ax2, title="해커스실전2_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()


figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
hackers_real3_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="해커스실전3_못푼개수", cmap="rainbow")
hackers_real3_total.plot(kind="bar", legend=True, ax=ax2, title="해커스실전3_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()

figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)
hackers_real4_nonreplied.plot(kind="bar", legend=True, ax=ax1, title="해커스실전4_못푼개수", cmap="rainbow")
hackers_real4_total.plot(kind="bar", legend=True, ax=ax2, title="해커스실전4_못푼개수+틀린개수", cmap="rainbow")
for p in ax1.patches:
    ax1.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for q in ax2.patches :
    ax2.annotate(str(q.get_height()), (q.get_x() * 1.005, q.get_height() * 1.005))
plt.show()
```
